```python
import pandas as pd
import numpy as np
```

```python
sheet_name = 'Page1_'
book_name = 'NASF State Data Extract.xlsx'
df_detail = pd.DataFrame()
for i in range(1, 12):
    df_detail = df_detail.append(pd.read_excel(book_name, sheet_name=sheet_name + str(i), skiprows=7))
```

```python
df_detail.columns
```

```python
df_detail = df_detail.drop(['Local Incident ID', 'Incident Name', 
                'Fire Containment Date', 'Fire Containment Time',
                'State FIPS', 'District',
                'County', 'County FIPS', 'Ownership Code', 
               ], axis = 1)
```

```python
summary = df_detail.loc[df_detail['Fire Discovery Date'].isna()].iloc[0,7:]
summary
df_detail = df_detail.dropna(subset=['Fire Discovery Date'])
```

```python
df_detail.loc[:, 'Year'] = pd.DatetimeIndex(df_detail.loc[:, 'Fire Discovery Date']).year
```

```python
df_detail.to_json('fire_data.json', orient='records')
```

```python
lost_per_year_per_state = df_detail.groupby(['Year', 'State']).sum().iloc[:, -6:]
```

```python
df = pd.read_json('fire_data.json', orient='records')
```

```python
lost_per_year_per_state = df.groupby(['Year', 'State']).sum().loc[:, ['Number Fatalities', 
                                                                      'Number Injuries',
                                                                      'Other Structures Destroyed',
                                                                      'Other Structures Threatened',
                                                                      'Residences Destroyed',
                                                                      'Residences Threatened',
                                                                      'Final Fire Acre Quantity'
                                                                     ]]
```

```python
lost_per_year_per_state.to_json('damage.json', orient='index')
```

```python
lost_per_year_per_state.loc[:,'Number of Fires'] = df.groupby(['Year', 'State']).count()
```

```python
lost_per_year_per_state.reset_index().groupby('Year').apply(lambda x: x.iloc[:,1:].to_dict(orient='records')).to_json('damage.json', orient='index')
```
